```python
import pandas as pd
import json
import pprint
import urllib.request

# Specify the file path or URL of the CSV file
file_path = "/Users/brennancurrie/Desktop/My_Code/My_Repos/group-project-3/Resources/spotify_data.csv"

# Read the CSV file and create a DataFrame
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Combine Track & Artist, and Popularity Column 
df['Track_Artist_Popularity'] = df.apply(lambda row: row['Track'] + ' - ' + row['Artist'] + ' (Popularity: ' + str(row['Popularity']) + ')', axis=1)
df
```

```python
#get average popularity per state
# Group the dataframe by the "state" column and calculate the average of the "popularity" column
avg_popularity_per_state = df.groupby('State')['Popularity'].mean()

avg_popularity_per_state_df = pd.DataFrame(avg_popularity_per_state)

# Print the average popularity by state
avg_popularity_per_state_df
```

```python
# Convert the DataFrame to JSON
pop_data = avg_popularity_per_state_df.to_json(orient='records')
```

```python
# Writing to json
with open("avg_pop_data.json", "w") as outfile:
    outfile.write(pop_data)
```

```python
# Drop the original columns 'Track', 'Artist', and 'Popularity' 
df.drop(['Track', 'Artist', 'Popularity'], axis=1, inplace=True)
df
```

```python
#Create Function to combine strings

def combine_fields(strings):
    return '\n'.join(strings)

#Groupby State
df = df.groupby(["State"]).agg(combine_fields).reset_index().copy()
df
```
